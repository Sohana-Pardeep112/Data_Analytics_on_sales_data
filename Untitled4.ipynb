```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
#dataset
sales_data = pd.read_csv('/content/retail_store_sales.csv')
print(sales_data.head())
print(sales_data.info())

#data cleaning
data_cleaned = sales_data.drop_duplicates()
print("Clean Data:\n",
      data_cleaned.head())

#missing values
data_missing_values = data_cleaned.isnull().sum()
print("Missing Values: \n" ,
      data_missing_values.head())

#handling missing values
#replacing missing values with mean
data_cleaned['Item'] = data_cleaned['Item'].fillna( 'Unknown'  )
data_cleaned.fillna({'Quantity' : data_cleaned['Quantity'].mean()}, inplace = True)
data_cleaned.fillna({'Price Per Unit' : data_cleaned['Price Per Unit'].mean()}, inplace = True)
data_cleaned.fillna({'Discount Applied' : data_cleaned['Discount Applied'].mode()[0]}, inplace = True)

#recalculating total spent
data_cleaned['Total Spent'] = data_cleaned['Price Per Unit']*data_cleaned['Quantity']
print(data_cleaned['Total Spent'])

#Standardize inconsistent region names
data_cleaned['Location'] = data_cleaned['Location'].str.strip().str.title()
print(data_cleaned['Location'].unique())

#Parse date formats correctly
data_cleaned['Transaction Date'] = pd.to_datetime(data_cleaned['Transaction Date'], errors = 'coerce')
data_cleaned['Month'] = data_cleaned['Transaction Date'].dt.to_period('M')
monthly_sales = data_cleaned.groupby('Month')['Total Spent'].sum().reset_index()



```

```python
#key metrices
total_sales = data_cleaned['Total Spent'].sum()
average_sales = data_cleaned['Total Spent'].mean()
sales_by_location = data_cleaned.groupby('Location')['Total Spent'].sum()

summary_data = {
    'Total Sales' : [total_sales],
    'Average_Sales' : [average_sales],
}
summary_file = pd.DataFrame(summary_data)


#save to excel
with pd.ExcelWriter('/content/cleaned_sales_summary.xlsx') as writer:
    data_cleaned.to_excel(writer, sheet_name='Cleaned Data', index=False)
    summary_file.to_excel(writer, sheet_name='Summary', index=False)
    sales_by_location.to_frame(name='Sales by Location').to_excel(writer, sheet_name='Sales by Location')
    monthly_sales.to_excel(writer, sheet_name='Monthly Sales', index=False)
```

```python
#BAR CHART
plt.bar(sales_by_location.index, sales_by_location.values, color = 'pink')
plt.title('Sales by Location')
plt.xlabel('Location')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Total Spent'], marker='o', linestyle='-', color='pink')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=90, fontsize=8)  # Adjust fontsize here

plt.show()
```

```python
sales_by_category = data_cleaned.groupby('Category')['Total Spent'].sum()

top_5_categories = sales_by_category.nlargest(5)
plt.pie(top_5_categories, labels=top_5_categories.index, autopct='%1.1f%%')
plt.title('Top 5 Categories by Sales')

plt.show()
```

```python
daily_sales = data_cleaned.groupby('Transaction Date')['Total Spent'].sum()
plt.boxplot(daily_sales.values)
plt.title('Boxplot of Daily Sales Amounts')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Calculate customer purchase frequency
customer_frequency = data_cleaned['Customer ID'].value_counts()

from matplotlib.ticker import MaxNLocator
# Create a histogram of customer purchase frequency
plt.hist(customer_frequency, bins=12, edgecolor='black', color = 'pink')

#maxlocator makes sure to show the ticks or use the data at integer number only
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.title('Distribution of Customer Purchase Frequency')
plt.xlabel('Number of Purchases')
plt.ylabel('Number of Customers')

plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

df = pd.read_excel("cleaned_sales_summary (5).xlsx")

df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

snapshot_date = pd.to_datetime("2025-01-19")

rfm = df.groupby('Customer ID').agg({
    'Transaction Date': lambda x: (snapshot_date - x.max()).days,
    'Transaction ID': 'nunique',
    'Total Spent': 'sum'
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])

kmeans = KMeans(n_clusters=4)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

print(rfm.head(10))
```
